```python
import re
import nltk
import spacy
from nltk.stem import PorterStemmer
from nltk.util import ngrams

# Download (run once)
nltk.download('punkt')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = """
Machine learning is transforming the world. Deep learning models are powerful.
Natural language processing enables machines to understand human language.
"""

# ----------------------------
# Step 1: Clean + Tokenize
# ----------------------------
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = nltk.word_tokenize(text)
    return tokens

tokens = preprocess(text)

# ----------------------------
# Step 2: Merge words (bigrams)
# ----------------------------
def merge_words(tokens, n=2):
    n_grams = list(ngrams(tokens, n))
    merged = ["_".join(gram) for gram in n_grams]
    return merged

merged_tokens = merge_words(tokens, n=2)

# ----------------------------
# Step 3: Stemming
# ----------------------------
stemmer = PorterStemmer()

def apply_stemming(words):
    return [stemmer.stem(word) for word in words]

stemmed_words = apply_stemming(merged_tokens)

# ----------------------------
# Step 4: Lemmatization
# ----------------------------
def apply_lemmatization(words):
    doc = nlp(" ".join(words))
    return [token.lemma_ for token in doc]

lemmatized_words = apply_lemmatization(merged_tokens)

# ----------------------------
# Output comparison
# ----------------------------
print("\nOriginal Tokens:\n", tokens[:20])

print("\nMerged Tokens (Bigrams):\n", merged_tokens[:20])

print("\nAfter Stemming:\n", stemmed_words[:20])

print("\nAfter Lemmatization:\n", lemmatized_words[:20])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample corpus (3 documents)
documents = [
    "machine learning is fun",
    "deep learning is powerful",
    "machine learning and deep learning"
]

# ----------------------------
# 1. TF-IDF with unigrams
# ----------------------------
vectorizer_uni = TfidfVectorizer(ngram_range=(1,1))
X_uni = vectorizer_uni.fit_transform(documents)

print("\n--- UNIGRAM FEATURES ---")
print(vectorizer_uni.get_feature_names_out())

print("\n--- UNIGRAM TF-IDF VECTORS (dense) ---")
print(X_uni.toarray())


# ----------------------------
# 2. TF-IDF with bigrams
# ----------------------------
vectorizer_bi = TfidfVectorizer(ngram_range=(2,2))
X_bi = vectorizer_bi.fit_transform(documents)

print("\n--- BIGRAM FEATURES ---")
print(vectorizer_bi.get_feature_names_out())

print("\n--- BIGRAM TF-IDF VECTORS (dense) ---")
print(X_bi.toarray())


# ----------------------------
# 3. TF-IDF with uni + bigrams
# ----------------------------
vectorizer_all = TfidfVectorizer(ngram_range=(1,2))
X_all = vectorizer_all.fit_transform(documents)

print("\n--- UNI + BIGRAM FEATURES ---")
print(vectorizer_all.get_feature_names_out())

print("\n--- UNI + BIGRAM TF-IDF VECTORS ---")
print(X_all.toarray())
```
